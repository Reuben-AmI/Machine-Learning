## Gradient Descent

Gradient descent is an optimization algorithm used to find the values of parameters (coefficients) of a function (f) that minimizes a cost function (cost).



Gradient descent is best used when the parameters cannot be calculated analytically (e.g. using linear algebra) and must be searched for by an optimization algorithm.

---

In this short notebook we will apply Linear Regression using Gradient Descent on Artifically created Data.

#### First we create the Data & Plot

---

```python
import matplotlib.pyplot as plt
import numpy as np

## Create artificial data

X = np.random.rand(200,1)
y = -5 + 1*X + np.random.rand(200, 1)

## solve the normal equation

X_b = np.c_[np.ones((200, 1)), X]
theta_best = np.linalg.inv(X_b.T @ X_b)@(X_b.T)@(y)

##plot artificial data

plt.scatter(X, y)
plt.title("Artifical Data ")
```

### Now Using Scikit-Learn

We will now show how to create the same model but using SciKitLearn

---

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
eta = 0.1
m_iterations = 1000
m = 100


theta_collection = []
theta = np.random.rand(2, 1)
theta_collection.append(theta)

for _ in range(m_iterations):
    gradients = 2/m * X_b.T @ (X_b @ theta - y)
    theta = theta - eta*gradients
    if _ == 10 or _ == 100 or _ == 999:
        theta_collection.append(theta)
```

```python
plt.scatter(X, y, marker = "*")

plt.plot(X, [x*theta_collection[1][1] + theta_collection[1][0] for x in X],
         color = "red",
        label = "iteration = 10")
plt.plot(X, [x*theta_collection[2][1] + theta_collection[2][0] for x in X],
         color = "orange",
        label = "iteration = 100")
plt.plot(X, [x*theta_collection[3][1] + theta_collection[3][0] for x in X],
         color = "green",
        label = "iteration = 1000")
plt.title("Atrificial Data")
plt.legend()
```

### We see the gradual improvement of the model the more we iterate.
